# Filtro Laplaciano

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Funo para aplicar o filtro Laplaciano com diferentes mscaras
def aplicarFiltroLaplaciano(imagem, mascara):
    return cv2.filter2D(imagem, -1, mascara)

# Funo para normalizar a imagem no intervalo [0, 255]
def normalizarImagem(imagem):
    return cv2.normalize(imagem, None, 0, 255, cv2.NORM_MINMAX)

# Carregar a imagem em escala de cinza
imagem = cv2.imread('lena.png', cv2.IMREAD_GRAYSCALE)

# Definir as mscaras Laplacianas
mascaraA = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])
mascaraB = np.array([[1, 1, 1], [1, -8, 1], [1, 1, 1]])
mascaraC = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
mascaraD = np.array([[0, -1, 0], [-1, 4, -1], [0, -1, 0]])

# Aplicar os filtros Laplacianos usando as diferentes mscaras
laplacianoA = aplicarFiltroLaplaciano(imagem, mascaraA)
laplacianoB = aplicarFiltroLaplaciano(imagem, mascaraB)
laplacianoC = aplicarFiltroLaplaciano(imagem, mascaraC)
laplacianoD = aplicarFiltroLaplaciano(imagem, mascaraD)

# Normalizar os resultados das imagens Laplacianas
laplacianoA = normalizarImagem(laplacianoA)
laplacianoB = normalizarImagem(laplacianoB)
laplacianoC = normalizarImagem(laplacianoC)
laplacianoD = normalizarImagem(laplacianoD)

# Salvar as imagens normalizadas
cv2.imwrite('laplacianoA.png', laplacianoA)
cv2.imwrite('laplacianoB.png', laplacianoB)
cv2.imwrite('laplacianoC.png', laplacianoC)
cv2.imwrite('laplacianoD.png', laplacianoD)

# Exibir os resultados em subplots
imagens = [laplacianoA, laplacianoB, laplacianoC, laplacianoD]
titulos = ['Laplaciano A', 'Laplaciano B', 'Laplaciano C', 'Laplaciano D']

plt.figure(figsize=(10, 8))
for i in range(len(imagens)):
    plt.subplot(2, 2, i + 1)
    plt.imshow(imagens[i], cmap='gray')
    plt.title(titulos[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

# Filtro Gradiente

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Funo para normalizar a imagem no intervalo [0, 255]
def normalizarImagem(imagem):
    return cv2.normalize(imagem, None, 0, 255, cv2.NORM_MINMAX)

# Carregar a imagem em escala de cinza
imagem = cv2.imread('lena.png', cv2.IMREAD_GRAYSCALE)

# Aplicar o filtro Sobel para calcular os gradientes horizontais e verticais
sobelX = cv2.Sobel(imagem, cv2.CV_64F, 1, 0, ksize=3)  # Gradiente horizontal (Sobel X)
sobelY = cv2.Sobel(imagem, cv2.CV_64F, 0, 1, ksize=3)  # Gradiente vertical (Sobel Y)

# Normalizar os resultados dos gradientes
sobelX = normalizarImagem(np.abs(sobelX))
sobelY = normalizarImagem(np.abs(sobelY))

# Salvar as imagens dos gradientes
cv2.imwrite('gradienteSobelX.png', sobelX)
cv2.imwrite('gradienteSobelY.png', sobelY)

# Exibir os resultados dos gradientes
imagens = [sobelX, sobelY]
titulos = ['Gradiente Horizontal (Sobel X)', 'Gradiente Vertical (Sobel Y)']

plt.figure(figsize=(8, 6))
for i in range(len(imagens)):
    plt.subplot(1, 2, i + 1)
    plt.imshow(imagens[i], cmap='gray')
    plt.title(titulos[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
