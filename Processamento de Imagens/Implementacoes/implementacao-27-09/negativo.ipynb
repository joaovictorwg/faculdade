```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def aplicar_negativo(imagem):
    # n m√°ximo de n√≠veis de cinza
    L = 256
    
    # Aplicar a transforma√ß√£o negativa
    imagem_negativa = L - 1 - imagem
    
    return imagem_negativa

# Carregar a imagem em escala de cinza
imagem = cv2.imread('imgs/image.png', cv2.IMREAD_GRAYSCALE)

if imagem is None:
    print("Erro ao carregar a imagem.")
else:
    # Aplicar a transforma√ß√£o negativa
    imagem_negativa = aplicar_negativo(imagem)

    # Mostrar informa√ß√µes da imagem original e da imagem negativa
    print(f"Dimens√µes da Imagem Original: {imagem.shape}")

    # Visualizar as imagens
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.title("Imagem Original (Escala de Cinza)")
    plt.imshow(imagem, cmap='gray', interpolation='nearest')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Imagem Negativa")
    plt.imshow(imagem_negativa, cmap='gray', interpolation='nearest')
    plt.axis('off')

    plt.show()

    # Salvar a imagem negativa
    cv2.imwrite('imgs/imagem_negativa.jpg', imagem_negativa)
```
