# Algoritmo 1

```python
import time
from tabulate import tabulate

comparacoes = 0
trocas = 0

def merge(A, left, mid, right):
    global comparacoes, trocas
    # Tamanho dos subarrays
    n1 = mid - left + 1
    n2 = right - mid

    # Criação dos subarrays L e R
    L = [0] * n1
    R = [0] * n2

    # Preencher subarrays L e R
    for i in range(n1):
        L[i] = A[left + i]
    for j in range(n2):
        R[j] = A[mid + 1 + j]

    # Adicionar infinito no final
    L.append(float('inf'))
    R.append(float('inf'))

    i = 0
    j = 0

    # Mescla os subarrays L e R de volta para A
    for k in range(left, right + 1):
        comparacoes += 1  
        if L[i] <= R[j]:
            A[k] = L[i]
            i += 1
        else:
            A[k] = R[j]
            j += 1
        trocas += 1  

def merge_sort(A, left, right):
    if left < right:
        mid = (left + right) // 2
        merge_sort(A, left, mid)
        merge_sort(A, mid + 1, right)
        merge(A, left, mid, right)

def read_list_from_file(file_path):
    with open(file_path, 'r') as file:
        return [int(line.strip()) for line in file if line.strip()]


file_paths = [
    "listas/lista_aleatoria_10000.txt",
    "listas/lista_inversa_10000.txt",
    "listas/lista_ordenada_10000.txt"
]

for file_path in file_paths:
  
    A = read_list_from_file(file_path)
    n = len(A)

    start_time = time.time()
    merge_sort(A, 0, n - 1)
    end_time = time.time()

    # Calcular o tempo em milissegundos
    execution_time = (end_time - start_time) * 1000  

    data = [
        ["Arquivo", file_path],
        ["Tempo de Execução (ms)", execution_time],
        ["Número de Comparações", comparacoes],
        ["Número de Trocas", trocas]
    ]


    print(tabulate(data, headers=["Descrição", "Valor"], tablefmt="grid"))
    print("Array ordenado:", A[:10], "...")  # Exibir os primeiros 10 elementos do array ordenado

    comparacoes = 0
    trocas = 0
```

# Algoritmo 2

```python
import time
import math
from tabulate import tabulate

def read_list_from_file(file_path):
    with open(file_path, 'r') as file:
        return [int(line.strip()) for line in file if line.strip()]

# Função Merge_M2, que realiza a mesclagem de duas sublistas com base nas subsequências crescentes/decrescentes
def Merge_M2(A, p, q, r, B):
    n1 = B[q] - B[p] 
    n2 = B[r] - B[q] 
        
    contL = 1 if A[B[q]] - A[B[p] + 1] > 0 else -1
    contR = 1 if A[B[r]] - A[B[q] + 1] > 0 else -1
    
    L = [0] * (n1 + 1)
    R = [0] * (n2 + 1)
    
    if contL == 1:
        for i in range(n1):
            L[i] = A[B[p] + i + 1]
    else:
        for i in range(n1):
            L[i] = A[B[q] - i]
    
    if contR == 1:
        for i in range(n2):
            R[i] = A[B[q] + i + 1]
    else:
        for i in range(n2):
            R[i] = A[B[r] - i]
    
    L[n1] = float('inf')
    R[n2] = float('inf')
    
    i = j = 0
    for k in range(n1 + n2):
        if L[i] <= R[j]:
            A[B[p] + k + 1] = L[i]
            i += 1
        else:
            A[B[p] + k + 1] = R[j]
            j += 1
    return A

# Função Sort que aplica o algoritmo de ordenação com base na identificação de subsequências
def Sort(A, n):
    B = [-1] * (n + 1)
    t = 0
    cont = -1
    
    if A[0] <= A[1]:
        cont = 1
    
    for i in range(n - 2):
        if cont * A[i + 1] < cont * A[i]:
            t += 1
            B[t] = i
            cont = -1
            if A[i + 1] <= A[i + 2]:
                cont = 1
    
    if cont * A[n - 1] < cont * A[n - 2]:
        t += 1
        B[t] = n - 2
    
    k = t + 1
    B[t + 1] = n - 1
    
    while k > 1:
        t = 0
        for i in range(0, k - 1, 2):
            A = Merge_M2(A, i, i + 1, i + 2, B)
            t += 1
            B[t] = B[i + 2]
        B[t + 1] = n - 1
        k = math.ceil(k / 2)
    
    return A

# Caminhos dos arquivos de listas
file_paths = [
    "listas/lista_aleatoria_10000.txt",
    "listas/lista_inversa_10000.txt",
    "listas/lista_ordenada_10000.txt"
]

# Processar cada arquivo de lista
for file_path in file_paths:
    A = read_list_from_file(file_path)
    n = len(A)

    # Contadores para número de comparações e trocas (só como exemplo, incrementos podem ser adicionados onde apropriado)
    comparacoes = 0
    trocas = 0

    # Medir o tempo de execução
    start_time = time.time()
    sorted_array = Sort(A, n)
    end_time = time.time()

    # Calcular o tempo em milissegundos
    execution_time = (end_time - start_time) * 1000  

    # Exibir o resultado em formato de tabela
    data = [
        ["Arquivo", file_path],
        ["Tempo de Execução (ms)", execution_time],
        ["Número de Comparações", comparacoes],
        ["Número de Trocas", trocas]
    ]

    print(tabulate(data, headers=["Descrição", "Valor"], tablefmt="grid"))
    print("Array ordenado:", sorted_array[:10], "...")
```
