```python
import networkx as nx
import matplotlib.pyplot as plt

# Definição do grafo
edges = [
    ("A", "B", 3), ("A", "C", 4), ("B", "D", 2), ("B", "E", 5),
    ("C", "F", 6), ("D", "G", 1), ("E", "G", 2), ("E", "H", 3),
    ("F", "H", 4), ("F", "I", 5), ("G", "J", 3), ("H", "J", 2),
    ("H", "K", 4), ("I", "K", 6), ("J", "L", 2), ("K", "L", 3),
    ("K", "M", 5), ("L", "N", 4), ("M", "N", 3), ("N", "O", 2),
]

# Criar o grafo e adicionar arestas com pesos
G = nx.DiGraph()
for u, v, w in edges:
    G.add_edge(u, v, weight=w)

# Definir os vértices inicial e final
start_node = "A"
end_node = "O"

# Resolver o caminho mais curto usando o algoritmo de Dijkstra
shortest_path = nx.shortest_path(G, source=start_node, target=end_node, weight="weight")
shortest_path_length = nx.shortest_path_length(G, source=start_node, target=end_node, weight="weight")

print("Caminho mais curto:", shortest_path)
print("Comprimento do caminho:", shortest_path_length)

pos = nx.spring_layout(G)  # Layout para o grafo
plt.figure(figsize=(12, 8))


nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=3000, font_size=10, font_weight="bold")
labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)


plt.title("Grafo", fontsize=15)
plt.show()

# Plotar o grafo completo
pos = nx.spring_layout(G)  # Layout para o grafo
plt.figure(figsize=(12, 8))

# Desenhar os nós e as arestas
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=2000, font_size=10, font_weight="bold")
labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

# Destaque o caminho mais curto
path_edges = list(zip(shortest_path, shortest_path[1:]))
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color="red", width=2)

plt.title("Grafo com Caminho Mais Curto Destacado", fontsize=15)
plt.show() 

```
